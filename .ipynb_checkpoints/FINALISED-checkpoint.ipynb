<div style="text-align: center; font-size: 24px; font-weight: bold;">
Levy-Type Solution for a Rectangular Plate (a × b) Under Sinusoidal Loading
</div>

<div style="text-align: center; font-size: 18px; font-weight: bold;">
Levy’s Method for Mixed Boundary Conditions
</div>

<br><br>

<p align="left"><b><font size="3">Loading Function</font></b></p>
<br>

$$ q(x,y) = q_0 \sin\left(\frac{\pi x}{a}\right) $$

This loading contains only a single sine component in the \( x \)-direction.

Therefore, the modal Fourier load (projection onto the \(y\)-direction sine basis) is

$$
q_n(x)
=
\frac{2}{b}
\int_0^b 
q_0 \sin\left(\frac{\pi x}{a}\right)
\sin\left(\frac{n\pi y}{b}\right)\, dy.
$$

Since the load is independent of \(y\),

$$
q_n(x)
=
q_0 \sin\left(\frac{\pi x}{a}\right)
\cdot
\frac{2}{b}
\int_0^b 
\sin\left(\frac{n\pi y}{b}\right)\, dy,
$$

and

$$
\int_0^b \sin\left(\frac{n\pi y}{b}\right)\,dy
=
\begin{cases}
\dfrac{2b}{n\pi}, & n \text{ odd},\\[6pt]
0, & n \text{ even}.
\end{cases}
$$

Thus the modal load becomes

$$
q_n(x)
=
\begin{cases}
\dfrac{4q_0}{n\pi}\sin\left(\dfrac{\pi x}{a}\right), & n \text{ odd}, \\[8pt]
0, & n \text{ even}.
\end{cases}
$$

Therefore only odd modes are forced.

<br><br>

<p align="left"><b><font size="3">Mode Expansion</font></b></p>
<br>

$$
w(x,y)
=
\sum_{n=1}^{\infty}
W_n(x)\,
\sin\left(\frac{n\pi y}{b}\right)
$$

Since \( y=0 \) and \( y=b \) are simply supported,

$$
w = 0, \qquad M_y = 0 \quad \text{at } y=0,b.
$$

<br><br>

<p align="left"><b><font size="3">Modal Governing ODE</font></b></p>
<br>

Substituting into the plate equation 
$$
\nabla^4 w = \frac{q}{D}
$$
gives

$$
W_n''''(x)
-
2\alpha_n^2 W_n''(x)
+
\alpha_n^4 W_n(x)
=
\frac{q_n(x)}{D}
$$

where

$$
\alpha_n = \frac{n\pi}{b}.
$$

<br><br>

<p align="left"><b><font size="3">Boundary Conditions in the x-Direction</font></b></p>
<br>

<b>Free edge at \( x = 0 \):</b>

$$
M_x(0)=0, \qquad Q_x(0)=0
$$

Plate relations:

$$
M_x = -D\left( W_n'' - \nu \alpha_n^2 W_n \right)
$$

$$
Q_x = -D\left( W_n''' - (2-\nu)\alpha_n^2 W_n' \right)
$$

So the BCs are:

$$
W_n''(0) - \nu\alpha_n^2 W_n(0) = 0
$$

$$
W_n'''(0) - (2-\nu)\alpha_n^2 W_n'(0) = 0
$$

<b>Clamped edge at \( x=a \):</b>

$$
W_n(a)=0, \qquad W_n'(a)=0
$$

<br><br>

<p align="left"><b><font size="3">Finite-Difference Approximation</font></b></p>
<br>

Let
$$
x_i = i\,\Delta x, \qquad i = 0,\dots,N-1.
$$


Fourth derivative:

$$
W''''(x_i)
\approx
\frac{
W_{i-2}-4W_{i-1}+6W_i-4W_{i+1}+W_{i+2}
}{\Delta x^4}
$$

Second derivative:

$$
W''(x_i)
\approx
\frac{
W_{i-1} - 2W_i + W_{i+1}
}{\Delta x^2}
$$

Interior finite-difference equation for
$$
i = 2,\dots,N-3
$$

$$
\frac{
W_{i-2}-4W_{i-1}+6W_i-4W_{i+1}+W_{i+2}
}{\Delta x^4}
-
2\alpha_n^2
\frac{
W_{i-1}-2W_i+W_{i+1}
}{\Delta x^2}
+
\alpha_n^4 W_i
=
\frac{q_n(x_i)}{D}
$$

<br><br>

<p align="left"><b><font size="3">Discrete Boundary Conditions</font></b></p>
<br>

<b>Moment-free at \( x=0 \):</b>

$$
W_n''(0)
\approx
\frac{
2W_0 - 5W_1 + 4W_2 - W_3
}{\Delta x^2}
$$

$$
\left(\frac{2}{\Delta x^2} - \nu\alpha_n^2\right)W_0
-\frac{5}{\Delta x^2}W_1
+\frac{4}{\Delta x^2}W_2
-\frac{1}{\Delta x^2}W_3
=0
$$

<b>Shear-free at \( x=0 \):</b>

$$
W_n'''(0)
\approx
\frac{
-W_0 + 3W_1 - 3W_2 + W_3
}{\Delta x^3}
$$

$$
W_n'(0)
\approx
\frac{
-3W_0 + 4W_1 - W_2
}{2\Delta x}
$$

So

$$
\frac{
-W_0 + 3W_1 - 3W_2 + W_3
}{\Delta x^3}
-
(2-\nu)\alpha_n^2
\frac{
-3W_0 + 4W_1 - W_2
}{2\Delta x}
=0
$$

<b>Clamped at \( x=a \):</b>

$$
W_{N-1} = 0
$$

$$
\frac{
3W_{N-1} - 4W_{N-2} + W_{N-3}
}{2\Delta x}
=0
$$

<br><br>

<p align="center"><b><font size="3">Modal Reconstruction</font></b></p>
<br>

$$
w(x,y)
=
\sum_{n=1}^{N_{\text{terms}}}
W_n(x)
\sin\left(\frac{n\pi y}{b}\right)
$$

Only odd \(n\) contribute when the load is sinusoidal in \(x\).

<br><br>


```python
#!/usr/bin/env python3
"""
Levy-mode plate solver with convergence check for number of Levy modes.
Fixed: handles pure-x sinusoidal load q(x,y)=q0*sin(pi*x/a) while keeping
Levy expansion in y (W_n are functions of x).
"""

import numpy as np
import matplotlib.pyplot as plt
import warnings

# ================================================================
# 1) PARAMETERS / INPUTS
# ================================================================
# geometry / material / load / numerical
a = 1.0        # plate length in x [m]
b = 1.0        # plate length in y [m]
h = 0.01       # thickness [m]
E = 210e9      # Young's modulus [Pa]
nu = 0.3       # Poisson's ratio
q0 = 1000.0    # load amplitude [N/m^2]

N_terms = 7    # initial number of Levy modes (may be overridden by convergence)
Nx = 300       # x-grid points for FD solve
Ny = 200       # y-grid points for reconstruction and plotting

save_figures = False   # set True to save contour images to files
out_prefix = 'plate_'  # file prefix if saving

# Choose load axis: 'x' means q(x,y)=q0*sin(pi*x/a) (constant in y).
# 'y' means q(x,y)=q0*sin(pi*y/b) (constant in x) — original behavior.
load_axis = 'x'

# ================================================================
# 2) COMPUTE FLEXURAL RIGIDITY
# ================================================================
def compute_flexural_rigidity(E, h, nu):
    """Return flexural rigidity D = E h^3 / [12 (1 - nu^2)]."""
    return E * h**3 / (12.0 * (1.0 - nu**2))

# ================================================================
# 3) BUILD GRIDS
# ================================================================
def build_grids(a, b, Nx, Ny):
    """Return x (1D), y (1D), dx, dy, and mesh X,Y for plotting."""
    x = np.linspace(0.0, a, Nx)
    y = np.linspace(0.0, b, Ny)
    dx = x[1] - x[0]
    dy = y[1] - y[0]
    X, Y = np.meshgrid(x, y)
    return x, y, dx, dy, X, Y

# ================================================================
# 4) ASSEMBLE & SOLVE LEVY MODES (per-mode FD solve in x)
#    Returns modal amplitudes W_modes (N_terms x Nx)
# ================================================================
def assemble_and_solve_levy_modes(a, b, Nx, N_terms, dx, D, q0, nu, load_axis='x', q_func=None):
    """
    Solve for W_n(x) for n=1..N_terms using finite-difference in x.
    Boundary conditions:
      x=0 : free (Mx=0, Qx=0)
      x=a : clamped (W=0, W'=0)

    load_axis: 'x' for q(x,y)=q0*sin(pi*x/a) (constant in y) — analytic projection used.
               'y' for q(x,y)=q0*sin(pi*y/b) (constant in x) — original behavior.
    q_func: optional callable q(x_scalar, y_array) used only if load_axis not 'x' or 'y' defaults.
    """
    W_modes = np.zeros((N_terms, Nx))
    x = np.linspace(0.0, a, Nx)

    # --- compute modal load q_n(x) ---
    def compute_qn(n, x_arr):
        # CASE 1: load purely sinusoidal in x (independent of y) -> analytic projection over y
        if load_axis == 'x':
            # q(x,y) = q0 * sin(pi*x/a)
            # projection over y: integral_0^b sin(n pi y / b) dy = 2b/(n pi) for n odd, 0 for n even
            if n % 2 == 0:
                return np.zeros_like(x_arr)
            else:
                coef = 4.0 * q0 / (n * np.pi)   # derived analytic coefficient
                return coef * np.sin(np.pi * x_arr / a)
        # CASE 2: load purely sinusoidal in y (original behavior)
        elif load_axis == 'y' and q_func is None:
            # q(x,y) = q0 * sin(pi*y/b) -> only n==1 nonzero, independent of x
            if n == 1:
                return q0 * np.ones_like(x_arr)
            else:
                return np.zeros_like(x_arr)
        # CASE 3: general q(x,y) provided via q_func -> numerical projection in y
        else:
            # numerical projection over y via trapezoidal rule
            Ny_quad = max(200, Ny)  # use Ny or at least 200
            yq = np.linspace(0.0, b, Ny_quad)
            sin_basis = np.sin(n * np.pi * yq / b)
            coef = 2.0 / b
            qn_x = np.zeros_like(x_arr)
            for i, xi in enumerate(x_arr):
                qy = q_func(xi, yq)
                qn_x[i] = coef * np.trapz(qy * sin_basis, yq)
            return qn_x

    def build_matrix(alpha):
        N = Nx
        A = np.zeros((N, N))
        # Interior rows (i = 2 .. N-3) using centered FD stencils
        for i in range(2, N - 2):
            # W'''' central 5-point: [1, -4, 6, -4, 1] / dx^4
            A[i, i - 2] += 1.0 / dx**4
            A[i, i - 1] += -4.0 / dx**4
            A[i, i]     += 6.0 / dx**4
            A[i, i + 1] += -4.0 / dx**4
            A[i, i + 2] += 1.0 / dx**4
            # -2 alpha^2 * W'' central 3-point: [1, -2, 1] / dx^2
            A[i, i - 1] += -2.0 * alpha**2 * (1.0 / dx**2)
            A[i, i]     += -2.0 * alpha**2 * (-2.0 / dx**2)
            A[i, i + 1] += -2.0 * alpha**2 * (1.0 / dx**2)
            # + alpha^4 W
            A[i, i] += alpha**4
        return A

    def apply_BC(A, rhs, alpha):
        N = Nx
        # Row 0: Mx(0) -> W''(0) - nu*alpha**2 * W(0) = 0
        A[0, :] = 0.0
        A[0, 0] = (2.0 / dx**2) - nu * alpha**2
        A[0, 1] = (-5.0 / dx**2)
        A[0, 2] = (4.0 / dx**2)
        A[0, 3] = (-1.0 / dx**2)
        rhs[0] = 0.0

        # Row 1: Qx(0) -> W'''(0) - (2-nu)*alpha**2 * W'(0) = 0
        A[1, :] = 0.0
        # W''' coefficients: (-1, 3, -3, 1)/dx^3
        A[1, 0] += (-1.0) / dx**3
        A[1, 1] += (3.0)  / dx**3
        A[1, 2] += (-3.0) / dx**3
        A[1, 3] += (1.0)  / dx**3
        # subtract (2-nu)*alpha^2 * W'(0), W'(0) ≈ (-3 W0 + 4 W1 - W2) / (2 dx)
        c = (2.0 - nu) * alpha**2
        A[1, 0] += - c * (-3.0) / (2.0 * dx)
        A[1, 1] += - c * (4.0)  / (2.0 * dx)
        A[1, 2] += - c * (-1.0) / (2.0 * dx)
        rhs[1] = 0.0

        # Row N-1: W(a) = 0 (clamped)
        A[-1, :] = 0.0
        A[-1, -1] = 1.0
        rhs[-1] = 0.0

        # Row N-2: W'(a) = 0 using backward 2nd-order
        A[-2, :] = 0.0
        A[-2, -1] = 3.0 / (2.0 * dx)
        A[-2, -2] = -4.0 / (2.0 * dx)
        A[-2, -3] = 1.0 / (2.0 * dx)
        rhs[-2] = 0.0

        return A, rhs

    # Solve each Levy mode
    for n in range(1, N_terms + 1):
        alpha = n * np.pi / b
        qn = compute_qn(n, x)
        rhs = qn / D

        A = build_matrix(alpha)
        A, rhs = apply_BC(A, rhs, alpha)

        try:
            W = np.linalg.solve(A, rhs)
        except np.linalg.LinAlgError:
            warnings.warn(f"Mode {n}: singular matrix — using least-squares solve.")
            W = np.linalg.lstsq(A, rhs, rcond=None)[0]
        W_modes[n - 1, :] = W

    return W_modes

# ================================================================
# 5) RECONSTRUCT FULL DEFLECTION w(x,y)
# ================================================================
def reconstruct_deflection(W_modes, x, y, N_terms, b):
    """
    Build w(x,y) = sum_n W_n(x) * sin(n*pi*y/b)
    W_modes is (N_terms x Nx)
    Returns w with shape (Ny, Nx)
    """
    Ny = y.size
    Nx = x.size
    w = np.zeros((Ny, Nx))
    Y = np.tile(y[:, np.newaxis], (1, Nx))  # shape (Ny, Nx)
    for n in range(1, N_terms + 1):
        Wn = W_modes[n - 1, :][np.newaxis, :]  # shape (1, Nx)
        w += Wn * np.sin(n * np.pi * Y / b)
    return w

# ================================================================
# 6) COMPUTE DERIVATIVES (first and second) — helper
# ================================================================
def compute_derivatives(x, y, w):
    """
    Compute dw/dx, dw/dy, d2w/dx2, d2w/dy2, d2w/dxdy using np.gradient.
    Returns a dict of derivative arrays (same shape as w).
    """
    dx = x[1] - x[0]
    dy = y[1] - y[0]

    dw_dx = np.gradient(w, dx, axis=1)
    dw_dy = np.gradient(w, dy, axis=0)

    d2w_dx2 = np.gradient(dw_dx, dx, axis=1)
    d2w_dy2 = np.gradient(dw_dy, dy, axis=0)
    d2w_dxdy = np.gradient(dw_dx, dy, axis=0)

    return {
        'dw_dx': dw_dx,
        'dw_dy': dw_dy,
        'd2w_dx2': d2w_dx2,
        'd2w_dy2': d2w_dy2,
        'd2w_dxdy': d2w_dxdy
    }

# ================================================================
# 7) COMPUTE BENDING MOMENTS AND TRANSVERSE SHEARS
# ================================================================
def compute_moments_and_shears(derivs, D, nu):
    """
    Using Kirchhoff plate relations:
      Mx  = -D (w_xx + nu * w_yy)
      My  = -D (w_yy + nu * w_xx)
      Mxy = -D (1-nu) w_xy
    """
    d2w_dx2 = derivs['d2w_dx2']
    d2w_dy2 = derivs['d2w_dy2']
    d2w_dxdy = derivs['d2w_dxdy']

    Mx  = -D * (d2w_dx2 + nu * d2w_dy2)
    My  = -D * (d2w_dy2 + nu * d2w_dx2)
    Mxy = -D * (1.0 - nu) * d2w_dxdy

    return {
        'Mx': Mx,
        'My': My,
        'Mxy': Mxy
    }

def compute_shears_from_moments(Mx, My, Mxy, x, y):
    """Compute Qx and Qy from moment fields with correct spacing using np.gradient."""
    dx = x[1] - x[0]
    dy = y[1] - y[0]

    dMx_dx  = np.gradient(Mx, dx, axis=1)
    dMy_dy  = np.gradient(My, dy, axis=0)
    dMxy_dx = np.gradient(Mxy, dx, axis=1)
    dMxy_dy = np.gradient(Mxy, dy, axis=0)

    Qx = - (dMx_dx + dMxy_dy)
    Qy = - (dMxy_dx + dMy_dy)

    return Qx, Qy

# ================================================================
# 8) PLOTTING: contour helper and main plotting function
# ================================================================
def plot_contour(x, y, field, title, cb_label=None, levels=60, figsize=(8,6), cmap='jet'):
    X, Y = np.meshgrid(x, y)
    plt.figure(figsize=figsize)
    cf = plt.contourf(X, Y, field, levels, cmap=cmap)
    cbar = plt.colorbar(cf)
    if cb_label:
        cbar.set_label(cb_label)
    plt.xlabel('x [m]')
    plt.ylabel('y [m]')
    plt.title(title)
    plt.gca().set_aspect('equal', 'box')
    plt.tight_layout()
    plt.show()

def plot_all_contours(x, y, w, moments, Qx, Qy, save_figures=False, out_prefix='plate_'):
    # Deflection
    plot_contour(x, y, w, 'Deflection w(x,y) [m]', cb_label='w [m]')
    if save_figures:
        plt.savefig(out_prefix + 'w.png', dpi=300)

    # Moments
    plot_contour(x, y, moments['Mx'], 'Bending moment Mx [N·m]', cb_label='Mx [N·m]')
    if save_figures:
        plt.savefig(out_prefix + 'Mx.png', dpi=300)

    plot_contour(x, y, moments['My'], 'Bending moment My [N·m]', cb_label='My [N·m]')
    if save_figures:
        plt.savefig(out_prefix + 'My.png', dpi=300)

    plot_contour(x, y, moments['Mxy'], 'Twisting moment Mxy [N·m]', cb_label='Mxy [N·m]')
    if save_figures:
        plt.savefig(out_prefix + 'Mxy.png', dpi=300)

    # Shears
    plot_contour(x, y, Qx, 'Transverse shear Qx [N/m]', cb_label='Qx [N/m]')
    if save_figures:
        plt.savefig(out_prefix + 'Qx.png', dpi=300)

    plot_contour(x, y, Qy, 'Transverse shear Qy [N/m]', cb_label='Qy [N/m]')
    if save_figures:
        plt.savefig(out_prefix + 'Qy.png', dpi=300)

# ================================================================
# CONVERGENCE CHECK WITH RESPECT TO NUMBER OF LEVY MODES
# ================================================================
def levy_mode_convergence(a, b, x, Nx, dx, D, q0, nu, y,
                          max_N=15, tol=1e-4, plot=True, verbose=True, load_axis='x', q_func=None):
    """
    Perform convergence check over N_terms (Levy modes).
    """
    N_list = []
    wmax_list = []
    rel_list = []

    if verbose:
        print("\n===== Levy-Mode Convergence Check =====")
        print(" N   max|w| [m]        rel_change")

    w_prev = None

    for N in range(1, max_N + 1):
        # Solve modes and reconstruct deflection
        Wm = assemble_and_solve_levy_modes(a, b, Nx, N, dx, D, q0, nu, load_axis=load_axis, q_func=q_func)
        wN = reconstruct_deflection(Wm, x, y, N, b)

        wmax = np.max(np.abs(wN))
        N_list.append(N)
        wmax_list.append(wmax)

        # compute relative change (safe guard on denom)
        if w_prev is None:
            rel = 0.0
        else:
            denom = w_prev if abs(w_prev) > 1e-16 else 1e-16
            rel = abs((wmax - w_prev) / denom)
        rel_list.append(rel)

        if verbose:
            print(f"{N:2d} {wmax: .6e}     {rel: .3e}")

        # auto-stop criterion
        if (w_prev is not None) and (rel < tol):
            if verbose:
                print(f"\nConvergence achieved at N = {N} (relative change {rel:.3e} < tol={tol})")
            # optional plot
            if plot:
                plt.figure(figsize=(6,4))
                plt.plot(N_list, wmax_list, marker='o', linestyle='-')
                plt.xlabel('Number of Levy modes (N_terms)')
                plt.ylabel('max |w| [m]')
                plt.title('Convergence of max deflection with N_terms')
                plt.grid(True)
                plt.tight_layout()
                plt.show()
            return N_list, wmax_list, rel_list, N

        w_prev = wmax

    if verbose:
        print("\nReached max_N without satisfying tolerance.")
    if plot:
        plt.figure(figsize=(6,4))
        plt.plot(N_list, wmax_list, marker='o', linestyle='-')
        plt.xlabel('Number of Levy modes (N_terms)')
        plt.ylabel('max |w| [m]')
        plt.title('Convergence of max deflection with N_terms')
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    return N_list, wmax_list, rel_list, max_N

# ================================================================
# MAIN: follow block-diagram sequence
# ================================================================
if __name__ == '__main__':
    # 1. compute D
    D = compute_flexural_rigidity(E, h, nu)

    # 2. build grids
    x, y, dx, dy, X_mesh, Y_mesh = build_grids(a, b, Nx, Ny)

    # 3. run convergence check to choose N_terms (override default)
    #    You can adjust max_N and tol as desired.
    N_list, wmax_list, rel_list, N_opt = levy_mode_convergence(
        a, b, x, Nx, dx, D, q0, nu, y,
        max_N=15, tol=1e-4, plot=True, verbose=True, load_axis=load_axis, q_func=None
    )

    print(f"\nRecommended/selected N_terms = {N_opt}\n")

    # 4. assemble & solve Levy modes -> modal shapes Wn(x)
    N_terms = N_opt
    W_modes = assemble_and_solve_levy_modes(a, b, Nx, N_terms, dx, D, q0, nu, load_axis=load_axis, q_func=None)

    # 5. reconstruct w(x,y)
    w = reconstruct_deflection(W_modes, x, y, N_terms, b)

    # 6. compute derivatives of w
    derivs = compute_derivatives(x, y, w)

    # 7. compute bending moments
    moments = compute_moments_and_shears(derivs, D, nu)  # returns Mx,My,Mxy

    # 8. compute shear resultants from the moments (with correct spacing)
    Qx, Qy = compute_shears_from_moments(moments['Mx'], moments['My'], moments['Mxy'], x, y)

    # 9. plot contours for deflection, moments and shears
    plot_all_contours(x, y, w, moments, Qx, Qy, save_figures=save_figures, out_prefix=out_prefix)

    # Print a short summary for your presentation slide
    max_w = np.max(np.abs(w))
    max_Mx = np.max(np.abs(moments['Mx']))
    max_Qx = np.max(np.abs(Qx))
    print(f"Computed fields: max|w| = {max_w:.6e} m, max|Mx| = {max_Mx:.6e} N·m, max|Qx| = {max_Qx:.6e} N/m")

    # Print modal ODE and reconstruction formula (text)
    print("\nModal ODE: D(d^4W_n/dx^4 - 2 alpha_n^2 d^2W_n/dx^2 + alpha_n^4 W_n) = q_n(x)")
    print("alpha_n = n*pi/b; for load_axis='x' and q(x,y)=q0*sin(pi*x/a),")
    print("q_n(x) = (4 q0/(n pi)) * sin(pi x / a) for n odd, and 0 for n even.")
    print("Reconstruction: w(x,y) = sum_{n=1}^N W_n(x) * sin(n*pi*y/b)")
```
